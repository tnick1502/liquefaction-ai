# Модели PPR (Pore Pressure Ratio)

Построение графиков для всех моделей из `src/math_models/ppr_models.py`.

```python
import sys
sys.path.insert(0, "..")

import numpy as np
from src.math_models.ppr_models import (
    cpt_pore_pressure_ma_wang,
    cpt_ppr_ma_wang,
    extended_cpt_ppr,
    logarithmic_ppr_raw,
    logarithmic_ppr_normalized,
)
from src.utils.plotter import plot_function
```

## 1. CPT-модель Ма и Ванга — поровое давление u(N)

```python
fig = plot_function(
    func=cpt_pore_pressure_ma_wang,
    func_params={"a": 0.5, "k": 0.05, "A": 100.0},
    x_range=(1, 500),
    x_arg_name="n_cycles",
    title="Ма и Ванг — поровое давление u(N)",
    x_label="N (число циклов)",
    y_label="u (кПа)",
    func_label="u(N) — CPT-модель",
)
fig.show()
```

## 2. CPT-модель Ма и Ванга — PPR(N)

```python
fig = plot_function(
    func=cpt_ppr_ma_wang,
    func_params={"a": 0.5, "k": 0.05, "A": 100.0, "sigma_3": 200.0},
    x_range=(1, 500),
    x_arg_name="n_cycles",
    title="Ма и Ванг — PPR(N)",
    x_label="N (число циклов)",
    y_label="PPR",
    func_label="PPR(N) = u(N) / σ₃",
)
fig.show()
```

## 3. Расширенная CPT-модель — PPR(N)

```python
fig = plot_function(
    func=extended_cpt_ppr,
    func_params={"a": 0.3, "b": 0.3, "k": 0.05, "A": 1.0, "n_max": 500.0},
    x_range=(1, 500),
    x_arg_name="n_cycles",
    title="Расширенная CPT-модель — PPR(N)",
    x_label="N (число циклов)",
    y_label="PPR",
    func_label="f₁(N) — расширенная CPT",
)
fig.show()
```

## 4. Ненормированная логарифмическая модель — g(N)

```python
fig = plot_function(
    func=logarithmic_ppr_raw,
    func_params={"A1": 2.0, "A2": 1.5, "k1": 0.1, "k2": 0.01},
    x_range=(1, 500),
    x_arg_name="n_cycles",
    title="Ненормированная логарифмическая модель g(N)",
    x_label="N (число циклов)",
    y_label="g(N)",
    func_label="g(N) = A₁·ln(k₁N+1) − A₂·ln(k₂N+1)",
)
fig.show()
```

## 5. Нормированная логарифмическая модель — PPR(N)

```python
fig = plot_function(
    func=logarithmic_ppr_normalized,
    func_params={"A": 1.0, "A1": 2.0, "A2": 1.5, "k1": 0.1, "k2": 0.01},
    x_range=(1, 500),
    x_arg_name="n_cycles",
    title="Нормированная логарифмическая модель PPR(N)",
    x_label="N (число циклов)",
    y_label="PPR",
    func_label="f₂(N) = C·g(N)",
)
fig.show()
```
