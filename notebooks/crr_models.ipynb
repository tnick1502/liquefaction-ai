# Модели CRR (Cyclic Resistance Ratio)

Построение графиков для всех моделей из `src/math_models/crr_models.py`.

```python
import sys
sys.path.insert(0, "..")

import numpy as np
from src.math_models.crr_models import (
    bilge_exponential_n_from_csr,
    bilge_exponential_csr,
    lentini_logarithmic_csr,
    guoxing_power_n_from_csr,
    guoxing_power_csr,
    meziane_logarithmic_csr,
    author_hyperbolic_csr,
)
from src.utils.plotter import plot_function
```

## 1. Экспоненциальная модель H. Bilge — N(CSR)

```python
fig = plot_function(
    func=bilge_exponential_n_from_csr,
    func_params={"alpha": 5.0, "beta": -8.0},
    x_range=(0.05, 0.5),
    x_arg_name="csr",
    title="H. Bilge — экспоненциальная модель N(CSR)",
    x_label="CSR",
    y_label="N",
    func_label="N = α·exp(β·CSR)",
    log_y=True,
)
fig.show()
```

## 2. Экспоненциальная модель H. Bilge — CSR(N)

```python
fig = plot_function(
    func=bilge_exponential_csr,
    func_params={"alpha": 5.0, "beta": -8.0},
    x_range=(1, 10000),
    x_arg_name="n_cycles",
    title="H. Bilge — экспоненциальная модель CSR(N)",
    x_label="N (число циклов)",
    y_label="CSR",
    func_label="CSR = ln(N/α) / β",
    log_x=True,
)
fig.show()
```

## 3. Логарифмическая модель V. Lentini — CSR(N)

```python
fig = plot_function(
    func=lentini_logarithmic_csr,
    func_params={"alpha": -0.03, "beta": 0.35},
    x_range=(1, 10000),
    x_arg_name="n_cycles",
    title="V. Lentini — логарифмическая модель CSR(N)",
    x_label="N (число циклов)",
    y_label="CSR",
    func_label="CSR = α·ln(N) + β",
    log_x=True,
)
fig.show()
```

## 4. Степенная модель C. Guoxing — N(CSR)

```python
fig = plot_function(
    func=guoxing_power_n_from_csr,
    func_params={"alpha": 1000.0, "beta": -2.5},
    x_range=(0.05, 0.5),
    x_arg_name="csr",
    title="C. Guoxing — степенная модель N(CSR)",
    x_label="CSR",
    y_label="N",
    func_label="N = α·CSR^β",
    log_y=True,
)
fig.show()
```

## 5. Степенная модель C. Guoxing — CSR(N)

```python
fig = plot_function(
    func=guoxing_power_csr,
    func_params={"alpha": 1000.0, "beta": -2.5},
    x_range=(1, 10000),
    x_arg_name="n_cycles",
    title="C. Guoxing — степенная модель CSR(N)",
    x_label="N (число циклов)",
    y_label="CSR",
    func_label="CSR = (N/α)^(1/β)",
    log_x=True,
)
fig.show()
```

## 6. Логарифмическая модель E. Meziane — CSR(N)

```python
fig = plot_function(
    func=meziane_logarithmic_csr,
    func_params={"alpha": 2.4, "beta": 9.5},
    x_range=(1, 10000),
    x_arg_name="n_cycles",
    title="E. Meziane — логарифмическая модель CSR(N)",
    x_label="N (число циклов)",
    y_label="CSR",
    func_label="CSR = (α − ln(N)) / β",
    log_x=True,
)
fig.show()
```

## 7. Гиперболическая модель авторов — CSR(N)

```python
fig = plot_function(
    func=author_hyperbolic_csr,
    func_params={"alpha": 0.2, "beta": 0.4},
    x_range=(1, 10000),
    x_arg_name="n_cycles",
    title="Авторская гиперболическая модель CSR(N)",
    x_label="N (число циклов)",
    y_label="CSR",
    func_label="CSR = β / N^(1−α)",
    log_x=True,
)
fig.show()
```
