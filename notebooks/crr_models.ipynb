# Модели CRR (Cyclic Resistance Ratio)

Графики по `src/math_models/crr_models.py`. Модели **CSR(N)** и **N(CSR)** имеют разные оси, поэтому ниже **два** объединённых графика: все зависимости CSR от N на одном и все N от CSR на другом.

```python
import sys
sys.path.insert(0, "..")

import numpy as np
from src.math_models.crr_models import (
    author_hyperbolic_csr,
    bilge_exponential_csr,
    bilge_exponential_n_from_csr,
    guoxing_power_csr,
    guoxing_power_n_from_csr,
    lentini_logarithmic_csr,
    meziane_logarithmic_csr,
)
from liquefaction_ai.utils import plot_curves_overlay
```

## CSR(N) — все модели на одном графике

```python
fig_csr = plot_curves_overlay(
    curves=[
        {
            "func": bilge_exponential_csr,
            "func_params": {"alpha": 5.0, "beta": -8.0},
            "label": "H. Bilge — CSR = ln(N/α)/β",
        },
        {
            "func": lentini_logarithmic_csr,
            "func_params": {"alpha": -0.03, "beta": 0.35},
            "label": "V. Lentini — CSR = α·ln(N)+β",
        },
        {
            "func": guoxing_power_csr,
            "func_params": {"alpha": 1000.0, "beta": -2.5},
            "label": "C. Guoxing — CSR = (N/α)^(1/β)",
        },
        {
            "func": meziane_logarithmic_csr,
            "func_params": {"alpha": 2.4, "beta": 9.5},
            "label": "E. Meziane — CSR = (α−ln(N))/β",
        },
        {
            "func": author_hyperbolic_csr,
            "func_params": {"alpha": 0.2, "beta": 0.4},
            "label": "Гипербола — CSR = β/N^(1−α)",
        },
    ],
    x_range=(1, 10000),
    x_arg_name="n_cycles",
    title="Сравнение моделей CSR(N)",
    x_label="N (число циклов)",
    y_label="CSR",
    log_x=True,
)
fig_csr.show()
```

## N(CSR) — все модели на одном графике

```python
fig_n = plot_curves_overlay(
    curves=[
        {
            "func": bilge_exponential_n_from_csr,
            "func_params": {"alpha": 5.0, "beta": -8.0},
            "label": "H. Bilge — N = α·exp(β·CSR)",
        },
        {
            "func": guoxing_power_n_from_csr,
            "func_params": {"alpha": 1000.0, "beta": -2.5},
            "label": "C. Guoxing — N = α·CSR^β",
        },
    ],
    x_range=(0.05, 0.5),
    x_arg_name="csr",
    title="Сравнение моделей N(CSR)",
    x_label="CSR",
    y_label="N",
    log_y=True,
)
fig_n.show()
```
