Сохраняет артефакт в каталог `data/` в корне репозитория (содержимое по умолчанию в git не коммитится, остаётся `.gitkeep`). Запуск: `poetry run jupyter lab`.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np

from liquefaction_ai import (
    generate_population,
    get_default_config,
    save_population_artifact,
    set_global_seed,
)

config = get_default_config()
# Для быстрой проверки можно уменьнить:
# config.n_scenarios = 4000
# config.benchmark_subset = 1500

set_global_seed(config.seed)
artifact_dir = Path("data/demo_run")
population = generate_population(config)
save_population_artifact(artifact_dir, population, config)
print("Сохранено в", artifact_dir.resolve())
```

### Краткая EDA

```python
meta = population["meta"]
fig, ax = plt.subplots(1, 2, figsize=(11, 4))
ax[0].hist(meta["D_r"], bins=30, color="#0b6efd", alpha=0.85)
ax[0].set_title("D_r")
ax[1].hist(population["liq_label"], bins=np.arange(-0.5, 2.5, 1), rwidth=0.7, color="#6610f2", alpha=0.85)
ax[1].set_title("liq_label (0/1)")
fig.tight_layout()
plt.show()

idx = 0
cycles = population["cycles"][idx]
plt.figure(figsize=(10, 4))
plt.plot(cycles, population["r_true"][idx], label="PPR true", lw=2)
plt.plot(cycles, population["r_obs"][idx], "--", label="PPR obs", alpha=0.8)
plt.xlabel("N")
plt.legend()
plt.title(f"Пример сценария idx={idx}")
plt.show()
```
